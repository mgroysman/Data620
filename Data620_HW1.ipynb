__Krackhardt Kite Graph:__

In graph theory, the Krackhardt kite graph is a simple graph with ten nodes. The graph is named after David Krackhardt, a researcher of social network theory.

Krackhardt introduced the graph in 1990 to distinguish different concepts of centrality. It has the property that the vertex with maximum degree (labeled 3 in the figure, with degree 6), the vertex with maximum betweenness centrality (labeled 7), and the two vertices with maximum closeness centrality (labeled 5 and 6) are all different from each other.[1]

[1] https://en.wikipedia.org/wiki/Krackhardt_kite_graph

__Python Code To Generate Krackhardt Kite Graph:__

Creating a graph using an available function.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.krackhardt_kite_graph()

nx.draw(G, with_labels=True)
plt.show()
```

Now I will attempt to build a graph from scratch.

```python
NG=nx.Graph()
```

Adding a node.

```python
NG.add_node('Andre')

nx.draw(NG,with_labels=True )
plt.show()
```

It will take a lot of work to do it one by one. Let's try to include them all in one shot.

```python
NG.add_nodes_from(['Beverly','Ed','Carol','Diane','Fernando','Ed','Garth','Heather','Ike','Jane'])

nx.draw(NG,with_labels=True )
plt.show()
```

We have nodes, but no edges. Let's add edges.

```python
NG.add_edge('Andre', 'Beverly')
nx.draw(NG,with_labels=True )
plt.show()
```

We have one edge. Let's add the rest of them.

```python
NG.add_edges_from([('Andre','Carol'),('Andre','Diane'),('Fernando','Andre'), ('Beverly','Ed'),('Beverly','Diane'),('Beverly','Garth'),
                  ('Ed','Diane'),('Ed','Garth'),('Garth','Diane'),('Garth','Fernando'),('Garth','Heather'),
                  ('Heather','Fernando'),('Heather','Ike'),('Ike','Jane'),('Fernando','Diane'),('Fernando','Carol'),
                 ('Carol','Diane')])
plt.figure(figsize=(8,8))
nx.draw(NG,node_size=400,with_labels=True )
plt.title("Krackhardt Kite Graph", fontsize=20)
plt.show()
```

Let's get some information about our graph.

```python
print(nx.info(NG))
```
